```python
def sum_even_numbers(start,end):
    _sum = 0 
    for i in range(start,end+1):
        if i%2==0:
            print(f'Adding {i} ..')
            _sum+=i
    return _sum
```

```python
def is_palindrome(string):
    _valid = list('qwertyuiopasdfghjklzxcvbnm') +list('QWERTYUIOPASDFGHJKLZXCVBNM')+ list('0123456789')
    i,j = 0,len(string)-1    
    while i<j:

        # skipping the left value if it is not in the _valid list 
        while (i<j) and (string[i] not in _valid):
            i+=1

        # skipping the right value if it is not in the _valid list 
        while (i<j) and (string[j] not in _valid):
            j-=1

        # comparing the left and right values
        if string[i].lower()==string[j].lower():
            i+=1
            j-=1
        else:
            return False

    return True
```

```python
print(is_palindrome("racecar"))                  # True
print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome("No lemon, no melon"))       # True
print(is_palindrome("hello"))                    # False
print(is_palindrome("Race a car"))                # False
```

```python
def factorial(n):
    if n<0:
        return 'negative value'
    if int(n)!=n:
        return 'decimal value'
    if n==0:
        return 1 
    else:
        return n*factorial(n-1)
```

```python
for i in range(0,10):
    print(factorial(i))
```

```python
def reverse_array(array):
    i,j = 0 ,len(array)-1
    while i<j:
        array[i],array[j] = array[j],array[i]
        i+=1
        j-=1
    return array
```

```python
def second_largest(list_):
    largest = list_[0]
    second = list_[1]

    for i in range(0,len(list_)):
        if largest<list_[i]:
            second = largest
            largest = list_[i]
            
        elif largest>list_[i]>second:
            second = list_[i]
    return second
```

```python
# A leap year is divisible by 4. Century years must also be divisible by 400 to be leap years.

def is_leap_year(year):
    if ((year%4==0) & (year%100!=0)) or (year%400==0):
        return True
    else:
        return False
```

```python
test_years = [2024, 2000, 2400, 1996, 1600, 1900, 2023, 2100, 2019, 1800]

for year in test_years:
    print(year, "->", is_leap_year(year))
```

```python
def count_vowels(string):
    vowels = 'aeiou'
    count = 0 
    for i in string:
        if i.lower() in vowels:
            count+=1
    return count 
```

```python
print(count_vowels("Hello World"))  # 3
print(count_vowels("AEIOU"))        # 5
print(count_vowels("Python"))       # 1
print(count_vowels("bcdfg"))        # 0
print(count_vowels(""))             # 0
```

```python
# Implement a function that merges two dictionaries
# {'name':'thomaskutty','age':29}, {'job':'data science','company':'google','age':29}

def merge_dictionaries(d1,d2):
    for key,value in d2.items():
        if key not in d1.keys():
            d1[key] = value
        else:
            pass
    return d1
            
```

```python
def longest_word(sentence):
    words = sentence.split()
    longest_word = words[0] 
    for i in range(len(words)):
        if len(words[i])>len(longest_word):
            longest_word = words[i]
        else:
            pass
    return longest_word
```

```python
def longest_word(sentence):
    max_length = 0
    start_index = 0      # start of current word
    longest_start = 0    # start of longest word found
    longest_end = 0      # end of longest word found

    for i in range(len(sentence)):
        if sentence[i] != ' ':
            # if it's the start of a word, mark start_index
            if i == 0 or sentence[i-1] == ' ':
                start_index = i

            # if end of sentence or next char is space, check word length
            if i == len(sentence) - 1 or sentence[i+1] == ' ':
                word_length = i - start_index + 1
                if word_length > max_length:
                    max_length = word_length
                    longest_start = start_index
                    longest_end = i

    return max_length, sentence[longest_start:longest_end+1]
```

```python
longest_word('i love mathematics and deep learning science')
```

```python
# 0,1,1,2,3,... 
def fibonacci(n):
    if n==0:
        return 0 
    elif n==1:
        return 1 
    else:
        return fibonacci(n-1)+fibonacci(n-2)
```

```python
for i in range(0,10):
    print(fibonacci(i))
```

```python
def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime

    for i in range(2, n):
        if n % i == 0:
            return False  # divisible → not prime
    return True  # no divisors → prime
```

```python
for i in range(2,10):
    print(i,is_prime(i))
```

```python
def gcd(a,b):
    output = 1 
    for i in range(2,min(a,b)+1):
        if ((a%i)==0 & (b%i)==0):
            output =i
    return output 
```

```python
def rotate_list(ar,k):
    output_array = [0]*len(ar)
    for i in range(len(ar)):
        output_array[(i+k)%len(ar)] = ar[i]
    return output_array
```

```python
rotate_list([23,4,5,8],2)
```

```python
def missing_number(ar):
    missing = [] 
    for i in range(len(ar)-1):
        diff = ar[i+1]-ar[i]
        if diff>1:
            for j in range(ar[i]+1,ar[i+1]):
                missing.append(j)
        else:
            pass
    return missing
```

```python
ar = [1, 2, 4, 7, 8]
print(missing_number(ar))
```

```python
def is_power_of_two(n: int) -> bool:
    if n < 1:
        return False  # 0 or negative numbers are not powers of two

    while n % 2 == 0:
        print(f'n = {n}')
        n = n//2  # divide n by 2 repeatedly

    return n == 1  # if we ended at 1, it was a power of two
```

```python
is_power_of_two(3)
```

```python
def sort(ar):
    wall = len(ar)-1
    while wall>0:
        for i in range(wall):
            if ar[i]>ar[i+1]:
                ar[i+1],ar[i] = ar[i],ar[i+1]
        wall-=1
        
    print(ar)
```

```python
# Check if a string has all unique characters
def is_unique_chars(string):
    repeat = {}
    for i in range(len(string)):
        if string[i] in repeat:
            return False
        else:
            repeat[string[i]]=None
    return True
```

```python
def max_subarray_sum(ar):
    max_subarray_val = float('-inf') 
    for i in range(len(ar)):
        # print(f'i={i}') 
        # print('----------')
        for j in range(i,len(ar)):
            # print(f'j={j}')
            # print('*********')
            sum_=0
            for k in range(i,j+1):
                # print(f'k={k}')
                sum_+=ar[k]
                
                
            if max_subarray_val<sum_:
                max_subarray_val = sum_
    return max_subarray_val
```

```python
def roman_to_integer(roman: str) -> int:
    # Map Roman numerals to their integer values
    roman_map = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    total = 0  # Result accumulator
    n = len(roman)
    
    for i in range(n):
        current_value = roman_map[roman[i]]
        
        # If this is not the last character and next numeral is larger, subtract
        if i < n - 1 and roman_map[roman[i]] < roman_map[roman[i + 1]]:
            total -= current_value
        else:
            total += current_value
    
    return total
```

```python
def are_anagrams_count(s1: str, s2: str) -> bool:
    """
    Check if two strings are anagrams by counting characters.
    Time complexity: O(n)
    """
    if len(s1) != len(s2):
        return False
    
    count1 = {}
    count2 = {}
    
    # Count characters for s1
    for ch in s1:
        count1[ch] = count1.get(ch, 0) + 1
    
    # Count characters for s2
    for ch in s2:
        count2[ch] = count2.get(ch, 0) + 1
    
    # Compare frequency dictionaries
    return count1 == count2
```
