# Map/Transform/Apply

```python
def seniority(row):
    if row['years_experience']>=6:
        return 'Senior'
    elif row['years_experience']>=4:
        return 'Mid'
    else:
        return 'Junior'

df['seniority'] = df.apply(seniority,axis=1)
```

```python
df[['employee_name','years_experience','seniority']]
```

```python
df[['salary', 'years_experience']].apply(
    lambda col: col / col.max()
)
```

```python
# For each column, get its type and number of missing values
def col_info(col):
    return pd.Series({
        'dtype': col.dtype,
        'num_missing': col.isna().sum(),
        'num_unique': col.nunique()
    })

df.apply(col_info, axis=0)
```

```python
# For numeric columns, compute mean, std, and range
def col_stats(col):
    if col.dtype in ['int64', 'float64']:
        return pd.Series({
            'mean': col.mean(),
            'std': col.std(),
            'range': col.max() - col.min()
        })

df.apply(col_stats, axis=0)
```

```python
# Add 10% bonus to salary
df['bonus'] = df['salary'].map(lambda x: x * 0.10)
```

```python
dept_map = {'Engineering': 'Eng', 'Finance': 'Fin', 'HR': 'HR', 'Marketing': 'Mkt'}
df['dept_short'] = df['department'].map(dept_map)
```

```python
df
```

```python
# Give a 5% salary hike to employees with experience > 5
df['salary'] = df['salary'] + 0.05 * df['salary'] * (df['years_experience'] > 5)
```

```python
df
```

```python
df['dept_avg_salary'] = df.groupby('department')['salary'].transform('mean')
df['above_avg'] = df['salary'] > df['dept_avg_salary']
```

```python
df
```

```python
# Round salary to nearest 1000 and add $500 bonus in one line
df['adjusted_salary'] = df['salary'].apply(lambda x: round(x, -3) + 500)
```

```python
# Percent difference from department average
df['pct_diff'] = df['salary'].transform(lambda x: (x - x.mean()) / x.mean() * 100)
```

# Concat - Merge - Join 

## concat : horizontal, vertical

```python
import pandas as pd

# First dataset
df1 = pd.DataFrame({
    'employee_id': [101, 102, 103],
    'employee_name': ['Alice', 'Bob', 'Charlie'],
    'department': ['Engineering', 'Engineering', 'HR']
})

# Second dataset
df2 = pd.DataFrame({
    'employee_id': [104, 105, 106],
    'employee_name': ['Diana', 'Ethan', 'Fiona'],
    'department': ['Finance', 'Marketing', 'Engineering']
})

display(df1)
display(df2)

# Vertical concatenation (stack rows)
df_vertical = pd.concat([df1, df2], ignore_index=True)
df_vertical
```

```python
df3 = pd.DataFrame({
    'salary':[90000,85000,60000]
})

pd.concat([df1,df3],axis=1)
```

## pd.merge 

```python
df_budget = pd.DataFrame({
    'department': ['Engineering', 'Finance', 'HR', 'Marketing'],
    'budget': [300000, 150000, 100000, 120000]
})

display(df_budget)

# Merge employee data with budget info
df_merged = pd.merge(df_vertical, df_budget, on='department', how='left')
df_merged
```

```python
# Suppose some departments are missing in budget
df_budget2 = pd.DataFrame({
    'department': ['Engineering', 'Finance'],
    'budget': [300000, 150000]
})

display(df_vertical)
display(df_budget2) 

# Left join (keep all employees)
display(pd.merge(df_vertical, df_budget2, on='department', how='left'))

# Inner join (only matching departments)
display(pd.merge(df_vertical, df_budget2, on='department', how='inner'))
```

```python
df_projects = pd.DataFrame({
    'department': ['Engineering', 'Engineering', 'HR'],
    'employee_name': ['Alice', 'Fiona', 'Charlie'],
    'project': ['Project X', 'Project Y', 'Project Z']
})
display(df_projects)
# Merge on department + employee_name
df_full = pd.merge(df_vertical, df_projects, on=['department', 'employee_name'], how='left')
df_full
```

## .join method ( how='left' ) : index based 

```python
# Set index to employee_id
df_vertical_indexed = df_vertical.set_index('employee_id')
df_salary = pd.DataFrame({
    'employee_id': [101, 102, 103],
    'salary': [90000, 85000, 60000]
}).set_index('employee_id')

display(df_vertical_indexed)
display(df_salary)

# Join on index
df_joined = df_vertical_indexed.join(df_salary,how='left')
df_joined
```

## pd.merge : Left_index , Right_index

```python
df_proj = pd.DataFrame({
    'project': ['Project X', 'Project Y', 'Project Z']
}, index=[101, 102, 104])  # 104 does not exist in df_emp

# Left join using index
pd.merge(df_emp, df_proj, left_index=True, right_index=True, how='left')
```

## pd.merge : Left_on, Right_on

```python
# Left DataFrame
df_emp = pd.DataFrame({
    'emp_id': [101, 102, 103],
    'employee_name': ['Alice', 'Bob', 'Charlie'],
    'department': ['Engineering', 'Engineering', 'HR']
})

# Right DataFrame
df_salary = pd.DataFrame({
    'employee_number': [101, 102, 103],
    'salary': [90000, 85000, 60000]
})

# Merge using left_on and right_on
df_merged = pd.merge(df_emp, df_salary, left_on='emp_id', right_on='employee_number', how='left')
df_merged
```

# Pivot Tables

```python

import pandas as pd

# Sample data
data = {
    "Date": ["2026-01-01", "2026-01-01", "2026-01-02", "2026-01-02"],
    "Product": ["A", "B", "A", "B"],
    "Region": ["North", "South", "North", "South"],
    "Sales": [100, 200, 150, 250]
}

df = pd.DataFrame(data)
display(df)

# Create a pivot table
pivot = pd.pivot_table(
    df, 
    values="Sales",        # Column to aggregate
    index="Product",       # Rows
    columns="Region",      # Columns
    aggfunc="sum",         # Aggregation function
    fill_value=0           # Replace NaN with 0
)

display(pivot)
```

```python
import pandas as pd

# Sample data
data = {
    "Date": ["2026-01-01", "2026-01-01", "2026-01-02", "2026-01-02", "2026-01-03", "2026-01-03"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Region": ["North", "South", "North", "South", "North", "South"],
    "Sales": [100, 200, 150, 250, 120, 300]
}

df = pd.DataFrame(data)
display(df)

# Custom function: range
def sales_range(x):
    return x.max() - x.min()

# Pivot table with multiple aggregation functions
pivot_multi = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    columns="Region",
    aggfunc=["sum", "max", sales_range],  # multiple aggregations
    fill_value=0
)

display(pivot_multi)
```

```python
import pandas as pd

# Sample data
data = {
    "Date": ["2026-01-01", "2026-01-01", "2026-01-02", "2026-01-02", "2026-01-03", "2026-01-03"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Region": ["North", "South", "North", "South", "North", "South"],
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Electronics", "Furniture"],
    "Sales": [100, 200, 150, 250, 120, 300]
}

df = pd.DataFrame(data)
display(df)

# Custom function
def sales_range(x):
    return x.max() - x.min()

# Pivot table with multiple rows and columns
pivot_multi_dim = pd.pivot_table(
    df,
    values="Sales",
    index=["Product", "Category"],   # multiple rows
    columns=["Region"],      # multiple columns
    aggfunc=["sum", sales_range],      # multiple aggregation functions
    fill_value=0
)

display(pivot_multi_dim)
```

```python
import pandas as pd

# Sample sales data
data = {
    "Product": ["A", "B", "A", "B", "C", "C"],
    "Region": ["North", "North", "South", "South", "North", "South"],
    "Sales": [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)
display(df)

# Pivot table with totals
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Product",
    columns="Region",
    aggfunc="sum",
    fill_value=0,
    margins=True,          # Add totals
    margins_name="Total"   # Name of the totals row/column
)

display(pivot)
```

# Time Series Functionalities 

```python
import pandas as pd

data = {
    "Date": ["2026-01-01 08:00", "2026-01-01 12:00", "2026-01-02 09:30", "2026-01-03 18:45"],
    "Sales": [100, 150, 200, 250]
}

df = pd.DataFrame(data)
display(df)

# Convert to datetime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Weekday"] = df["Date"].dt.weekday  # 0=Monday, 6=Sunday
df["Hour"] = df["Date"].dt.hour

display(df)
```

```python
# Set Date as index
df.set_index("Date", inplace=True)

# Daily sum of sales
daily_sales = df["Sales"].resample("D").sum()
daily_sales
```

```python
# Weekly sum
weekly_sales = df["Sales"].resample("W").sum()
weekly_sales
```

```python
# 2-day rolling sum
df['rolling_sum']= df["Sales"].rolling(window=2).sum()

# Expanding sum (cumulative sum)
df['expanding_sum'] = df["Sales"].expanding().sum()
df
```

```python
# | Code | Meaning                 | Example          |
# | ---- | ----------------------- | ---------------- |
# | `%d` | Day of month (0131)    | `01`, `10`, `31` |
# | `%m` | Month as number (0112) | `01`, `12`, `04` |
# | `%y` | 2-digit year            | `23`  2023      |
# | `%Y` | 4-digit year            | `2023`           |
# | `%H` | Hour (0023)            | `08`, `18`       |
# | `%M` | Minute (0059)          | `00`, `45`       |
# | `%S` | Second (0059)          | `00`, `30`       |
```

```python
import pandas as pd

data = {
    "Day": [1, 2, 3, 4],
    "Sales": [100, 150, 200, 250]
}

df = pd.DataFrame(data)


# Lag of 1 period
df["Sales_Lag1"] = df["Sales"].shift(1)

# Lead of 1 period
df["Sales_Lead1"] = df["Sales"].shift(-1)

display(df)
```

```python
data = {
    "Region": ["North", "North", "North", "South", "South", "South"],
    "Day": [1, 2, 3, 1, 2, 3],
    "Sales": [100, 150, 200, 50, 80, 120]
}

df = pd.DataFrame(data)

# Lag per region
df["Lag1"] = df.groupby("Region")["Sales"].shift(1)
df["Lead1"] = df.groupby("Region")["Sales"].shift(-1)

display(df)
```
