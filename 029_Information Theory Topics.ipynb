**Surprise - Self Information**

- Self Information - Surprisal
- How surprised should I be if this event happens?
- inversely proportional to probability

$$
\Large I(x) = -\log p(x)
$$

**Entropy** 

Entropy is the minimum average number of bits needed to describe the outcome of a random variable. Entropy is the expected surprise. 

$$
\Large H(X) = -\sum_x p(x)\log p(x)
$$

If two variables are independent, then $H(X,Y) = H(X) + H(Y)$.The maximum value of entropy is log n  which occurs when distribution is uniform. Minimum value is 0 when the outcome is certain. We can define conditional entropy $H(Y|X)$ which tells us how certain Y is even after knowing X. 

**Joint Entropy**

Total uncertainty of two random variables taken together. 

How many bits do I need, on average, to describe both X and Y? 

This is simply a direct extension of entropy. 

$$
\Large H(X,Y) = -\sum_{x,y} p(x,y)\log p(x,y)
$$

Follows chain rule; 

$$
\Large H(X,Y) = H(X) + H(Y|X)
$$

If X and Y are dependent; 

$$
H(X,Y) < H(X) + H(Y)
$$

**Conditional Entropy**

$$
H(Y|X) = -\sum_{x,y} p(x,y)\log p(y|x)
$$

**Mutual Information**

Redundancy between X and Y. 

$$
I(X;Y) = \sum_{x,y} p(x,y)\log\frac{p(x,y)}{p(x)p(y)}
$$

Cross Entropy

$$
H(p,q) = -\sum_x p(x)\log q(x)
$$

KL Divergence ( Relative Entropy) 

$$
KL(p||q) = \sum_x p(x)\log\frac{p(x)}{q(x)}
$$

Binary Cross Entropy

$$
L = -[y\log p + (1-y)\log(1-p)]
$$

Categorical Cross Entropy

$$
L = -\sum_i y_i\log p_i
$$
