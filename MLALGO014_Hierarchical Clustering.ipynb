Hierarchical clustering builds a **tree-like structure (dendrogram)** of the data points, showing how points/ clusters are merged (or split) at different distances.

- Two main types:
    1. **Agglomerative (Bottom-Up)** – start with each point as its own cluster, merge iteratively
    2. **Divisive (Top-Down)** – start with all points in one cluster, split iteratively

Most commonly used: **Agglomerative Clustering**

- **Distance Metric**: measure of dissimilarity between points : Euclidean, Manhattan, Cosine, etc.
- **Linkage Criteria**: measure of distance between clusters
    - **Single Linkage**: distance between closest points of clusters
    - **Complete Linkage**: distance between farthest points
    - **Average Linkage**: average distance between points in clusters
    - **Ward’s Method**: minimize variance within clusters

**Algorithm**

- **Start**: Each data point is its own cluster.
- **Compute Distances**: Calculate distances between all clusters.
- **Merge**: Find the two closest clusters and merge them.
- **Update Distances**: Recompute distances between the new cluster and all other clusters (based on chosen linkage).
- **Repeat**: Until all points are in **one cluster** or until **stopping criterion** (e.g., desired number of clusters or distance threshold) is reached

**Wards Method - Discussion**

- Instead of just looking at distance between points, it **looks at the increase in total within-cluster variance** if two clusters are merged.
- At each step, it merges the **pair of clusters that causes the smallest increase in total variance**.
