- DBSCAN groups together points that are **densely packed** and marks points in **low-density regions as noise/outliers**.
- Unlike K-Means:
    - **No need to specify the number of clusters** beforehand.
    - Can find **arbitrary-shaped clusters**

- Epsilon : maximum distance between two points for them to be considered neighbours.
- Minimum Points : Minimum number of points required in an epsilon neighbourhood to consider a point as core.
- Core Point : if we have enough number of points in the epsilon region.
- Border Point : If we donâ€™t have enough number of points (<minimum points) in the epsilon region.
- Noise Point : Neither core nor border

**Algorithm**

- Choose a point p that is not yet visited.
- Find neighbours within epsilon region.
- Classify the point:
    - If core point then start a cluster
    - Else temporarily label as noise ( might be border later)
- Expand the cluster
- Repeat for all points until all are visited.
